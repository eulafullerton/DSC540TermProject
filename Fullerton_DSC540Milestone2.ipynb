## 1. Combine Files

Use folder path to merge all the necessary data in order join all of the necessary txt files needed for analysis

```python
import os

# Specify the folder containing the .txt files
folder_path = 'Project data sets'  # Replace with the actual folder path

# Get a list of all .txt files in the folder
txt_files = [file for file in os.listdir(folder_path) if file.endswith('.txt')]

# Create an empty list to store DataFrames
df_list = []

# Loop through each file and read it into a DataFrame
for file in txt_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, delimiter='\t')  
    df_list.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# Display the first few rows of the combined DataFrame
combined_df.head()
```

```python
# Check original Census Regions 

# Get unique values in the 'Census Region' column
unique_census_regions = combined_df['Census Region'].unique()

# Display the unique values
print("Unique Census Regions:")
print(unique_census_regions)
```

## 2. Remove Unnecessary Rows

The combined data includes unnecessary rows that show information not needed for analysis

```python
# Check for missing values in each column
missing_values = combined_df.isna().sum()

# Display columns with missing values
print(missing_values)
```

```python
# Remove Rows with Missing or Empty Census Region
# Remove rows where "Census Region" is NaN
df_cleaned = combined_df.dropna(subset=['Census Region'])

# Display the cleaned DataFrame
df_cleaned.head()
```

```python
# check NaN again of cleaned df
cleaned_NaN = df_cleaned.isna().sum()
# Display columns with missing values
print(cleaned_NaN)
```

## 3. Remove unnecessary columns

The combined data includes unnecessary columns for analysis, therefore they are removed.

```python
# Remove whitespace from column names
df_cleaned.columns = df_cleaned.columns.str.strip()

# Specify the columns to remove
columns_to_remove = ['Chlamydia Code', 'Eclampsia Code', 'Gestational Diabetes Code', 
                     'Gestational Hypertension Code', 'Gonorrhea Code', 'Hepatitis B Code', 
                     'Hepatitis C Code', 'Pre-pregnancy Diabetes Code', 
                     'Pre-pregnancy Hypertension Code', 'Previous Preterm Birth Code', 
                     'Syphilis Code', 'Tobacco Use Code', 'Cause of death Code']

# Check which columns exist and drop them
existing_columns_to_remove = [col for col in columns_to_remove if col in df_cleaned.columns]
df_cleaned2 = df_cleaned.drop(columns=existing_columns_to_remove)
```

```python
# Display the NaN of the cleaned df
print(df_cleaned2.isna().sum())
```

## 3. Deal with missing values

Since the data is compiled from different extraction processes from the CDC natality data, missing values can be replaced with unknown in order to not get rid of necessary columns for the analysis. 

```python
# List of columns to replace missing values with "unknown"
columns_to_fill = ['Chlamydia', 'Eclampsia', 'Gestational Diabetes', 
                   'Gestational Hypertension', 'Gonorrhea', 'Hepatitis B', 
                   'Hepatitis C', 'Pre-pregnancy Diabetes', 
                   'Pre-pregnancy Hypertension', 'Previous Preterm Birth', 
                   'Syphilis', 'Tobacco Use', 'Cause of death']

# Fill missing values in the specified columns with "unknown"
df_cleaned2[columns_to_fill] = df_cleaned2[columns_to_fill].fillna("unknown")

# display new missing values
print(df_cleaned2.isna().sum())
```

```python
# Print the first five rows of df_cleaned2
print(df_cleaned2.head())
```
